```
len(df.movieId.unique())
```

```
user_movie_df=df.pivot(index='userId',columns='movieId',values='rating') #we are interchanging rows and columns because we are finding item based similarity
```

```
user_movie_df.iloc[0:5,0:10]
```

```
user_movie_df.fillna(0,inplace=True)
```

```
user_movie_df
```

```
from sklearn.metrics import pairwise_distances

user_sim=1-pairwise_distances(user_movie_df.values,metric='cosine') #cosine gives opposite answer , so we use 1-
```

```
import pandas as pd

#convert it to df for better understanding 
user_sim_df=pd.DataFrame(user_sim,index=df.userId.unique(),columns=df.userId.unique())
```

```
user_sim_df
```

```
#all diagonal user is similar to itself , so it is of no use

import numpy as np

np.fill_diagonal(user_sim,0)
```
